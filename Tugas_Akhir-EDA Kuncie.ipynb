## K-Means Cluster Analysist

K-means clustering dibuat untuk mengetahui data berdasarkan unsupervised modeling dengan penerapan pada elbow method yang akan dilakukan untuk mengetahui kluster-kluster data yang akan tervisualisasi berdasarkan plot-plot tersebut

#### Masukan import Library Pythonnya dahulu

```python
#import library pandas
import pandas as pd

#import library NumPy
import numpy as np

#import Plot Math Library
import matplotlib.pyplot as plt

#import Seaborn >> KMeans calculating
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn import preprocessing

%matplotlib inline
```

#### Masukan data Customer yang akan dianalisis. Studi kasus Data Customer pada Layanan Telekomunikasi

```python
raw_data = pd.read_csv('Data_tugasakhir_Kuncie.csv')
raw_data.head(10)
```

#### Membuat Index pada Gender, dan me-replace isi nya

```python
data = raw_data.set_index(['CustomerID'])
```

```python
mapping = {"Gender" : {"Male":0, "Female":1}}
data.replace(mapping, inplace=True)
```

```python
data.head(10)
```

```python
data.describe()
```

```python
data.nunique()
```

```python
data.isnull().sum()
```

### Merancang Plotting Data

```python
sns.pairplot(data)
```

### Merancang data berdasarkan Elbow Method untuk mengidentifikasi grafik pada sistem clustering

```python
#Merancang Elbow Method grafik dengan data full
# Initialize empty list of cost function values
wcss = []

# Create K-means models for different values of K
for i in range(2, 11):
    kmeans = KMeans(i)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_)  # inertia_ attribute gives the SSE value
    
# Plot the elbow graph
plt.plot(range(2, 11), wcss)
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.title('Elbow Method Graph')
plt.show()
```

### Skor Siluet untuk mengidentifikasi angka pada kluster

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans

sil = []
#Angka kluster dalam jangkauan 1-20
K = range(2, 11)
for n in K:
    algorithm = (KMeans(n_clusters = n) )
    algorithm.fit(data)
    labels = algorithm.labels_
    sil.append(silhouette_score(data, labels, metric = 'euclidean'))
```

```python
sil
```

Nilai ideal untuk menganalisis prediksi, dapat disimpulkan terbaik posisi range antara 4 atau 5

### Graph skor siluete sesuai dengan klustering

setelah didapatkan skoring siluete pada posisi range terbaik antara 4 atau 5. maka pembuktian selanjutnya yaitu dengan ditampilkan grafik skoring siluete apakah posisi yang dipilih telah pas ataupun tidak. sesuai dari analisis skoring tertinggi diantara nilai-nilai prediksi yang telah diberikan

```python
# Plot the silhouette scores
import matplotlib.pyplot as plt
plt.plot(K, sil)
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette score')
plt.title('Silhouette Scores for K-means Clustering')
plt.show()
```

### Analisis perbandingan antara Length of Stay dengan Telco Spending

Analisis Telco Spending dengan Length of Stay bertujuan untuk mengetahui cluster dari jumlah spending customer berdasarkan tingkat loyalitas yang didukung dengan jumlah rata-rata usia customers dan untuk mengetahui tingkatan customer behaviour terhadap produk yang ditawarkan oleh Penyedia Jasa Layanan Telekomunikasi.

```python
kmeans = KMeans(4)
kmeans.fit(data)
clusters = data.copy()
clusters['prediction'] = kmeans.fit_predict(clusters)
clusters.head(10)
```

```python
plt.scatter(data['Telco Spending (IDR 000)'],data['Length of Stay (months)'], c= clusters['prediction'], cmap='rainbow')
plt.xlabel('Telco Spending (IDR 000)')
plt.ylabel('Length of Stay (months)')
plt.title('K-means Clustering Telco Spending x Length Stay')
```

terlihat grafik persebaran kluster diatas terdapat 4 klustering yang terbagi setelah plot disusun antara perbandingan dari data telcospending dan length of stay

### Mengetahui Rata-rata nilai pada setiap warna clustering

```python
clusters.groupby('prediction').mean()
```

```python
# Group the data by 'prediction' column and calculate the mean of each column
grouped_data = clusters.groupby('prediction').mean()

# Create a new table with selected columns
new_table = grouped_data[['Age', 'Telco Spending (IDR 000)', 'Length of Stay (months)']]

# Print the new table
print(new_table)
```
